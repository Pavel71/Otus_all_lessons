```python
# Подсчитаем и выведем основных оптовиков для отдельной работы с ними
```

```python
def top_clien_id(df, procent_income):
    
    # Общее число клиентов
    client_size = len(df['user_id'].value_counts())
    
    # Общая выручка за указанный период 
    price_size = df['price'].sum()
    
    # Список клиентов в процентном соотношении
    df_top = df.groupby('user_id')['price'].sum().sort_values(ascending = False) / price_size *100
    
    number_clients = df_top.cumsum().searchsorted(procent_income)[0]+1
    
    procent_clients = round(number_clients / client_size *100,2)
    
    print('Кол-во клиентов = {0}({1}% из {2}) обеспечивающих {3}% от общей выручки'.format(
                                                                                   number_clients,
                                                                                   procent_clients,
                                                                                   client_size,
                                                                                   procent_income))
    
    Order_opt = pd.DataFrame()

    for number in df_top.iloc[:number_clients].index.values:
    
         Order_opt = Order_opt.append(Order[Order['user_id'] ==  number])
        
    # Делаем DataFrame с оптовыми покупаетлями    
    return   Order_opt 
```

```python
# Получаем DataFrame с оптовиками обеспечивающими 30% от всей выручки магазина с долей можно поигарть в зависимости
# от интереса магазина

OPT_client = top_clien_id(Order,30)
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import Imputer, FunctionTransformer
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.base import TransformerMixin
from sklearn.pipeline import make_union, make_pipeline


class Claster_pipeline(BaseEstimator):
    
    def __init__(self,df):
        self.df = df
        
    def discount_replace(self,df):
        
        df['DISCOUNT_VALUE'] = df['DISCOUNT_VALUE'].apply(lambda x: str(x).replace('%','')) 
        return np.vstack(df['DISCOUNT_VALUE'])
    
    def unix_time(self,df):
        
        # Чтобы работать с временными данными как с числами, преобразуем их к UNIX-time
        df['unix_time'] = df['date_insert'].astype(np.int64)/1000000000
        
        return np.vstack(df['unix_time'])
        
        
    def get_num_cols(self,df):
        return df[['PRICE', 'QUANTITY', 'discount', 'ordered_item_count', 'price']]

    def get_not_modified_cols(self,df):
        return df[['Name_clear', 'date_insert', 'itemID', 'orderID', 'user_id']]
    
    def get_pipline(self):

        pipeline = make_union(*[
            
            make_pipeline(FunctionTransformer(self.unix_time, validate=False),StandardScaler()),
            make_pipeline(FunctionTransformer(self.discount_replace, validate=False), Imputer(strategy='mean'), StandardScaler()),
            make_pipeline(FunctionTransformer(self.get_num_cols, validate=False), Imputer(strategy='mean'), StandardScaler())
#             make_pipeline(FunctionTransformer(self.get_not_modified_cols, validate=False))
        ])

        return pipeline.fit_transform(self.df)
```

```python
Claster_pip = Claster_pipeline(Order).get_pipline()
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

kmeans = KMeans(n_clusters=3, random_state=1)

kmeans.fit(Claster_pip)
clusters = kmeans.predict(Claster_pip)
```

```python
Order['cluster'] = clusters
```

```python
Order['cluster'].value_counts().plot(kind='bar');
```

```python

from mpl_toolkits.mplot3d import Axes3D

# Визуализируем получившееся разбиение на кластеры

fig = plt.figure(1, figsize=(10, 8))
ax = Axes3D(fig)

pca = PCA(n_components=3)
res = pca.fit_transform(Claster_pip)

ax.scatter(res[:, 0], res[:, 1], res[:, 2],
           c=kmeans.labels_, edgecolor='k')

ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')

ax.set_title('K-means clustering')
ax.dist = 12
```
