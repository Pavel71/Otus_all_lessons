```python
def grid_cv(model, param_grid, x_train, y_train):
    
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='neg_mean_squared_error')
    t_start = time.time()
    grid_search.fit(x_train, y_train)
    t_end = time.time()
    
    print('model {} best metric is {}'.format(model.__class__.__name__, grid_search.best_score_))
    print('time for training is {} seconds'.format(t_end - t_start))
    return grid_search.best_estimator_
```

```python
from sklearn.preprocessing import  LabelEncoder, LabelBinarizer, OneHotEncoder
from sklearn.pipeline import make_union, make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, Imputer
from sklearn.preprocessing import CategoricalEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import time

class Simple_pipline:
    
    '''
    Класс отвечает за преобразование данных
    Принимает таблицу с тренеровочной выборкой
    Преобразует важные признаки для модели обучения
    '''
    
    def __init__(self, df_ml, predict = False):
        
        '''
        При создание класса подгрузим таблицу для обучения из файла
        
        '''
        # Если нам нужен только предикт то обучаемся на всей выборке! Здесь нужно доработать
        if predict:
            
            self.df = pd.read_csv('simple_train_data.csv')
            
        else:
            
            self.df = df_ml
        
    def get_category(self,df):
        '''
        Функция возвращает категориальные признаки для обработки CategorialEncoding
        
        '''
        
        return df[['Прием пищи','Категория продукта','Оценка продукта','Год', 'Месяц', 'День',
       'День_недели']]
    
    def get_float_future(self,df):
        '''
        Функция возвращает числовые данные которые будут учавствовать в обучении
        Будет обрбаотка StandartScaller
        
        '''
        # Сюда я попробую подобавлять еще признаки по продуктам
        # Глюкоза
        return np.vstack(df['Углеводы'])
    
    def get_binary_future(self,df):
        '''
        Функция добавляет бинарные признаки! Пол, Образовательное учереждение, Тренировка, Болезнь 
        
        '''
        
        return np.vstack(df['Детский сад'])
    
    def pipline(self):
        '''
        Функция запускает обработчки pipline!
        
        '''
        
        pipline = make_union(*[
            
            make_pipeline(FunctionTransformer(self.get_category, validate=False),CategoricalEncoder(encoding='ordinal')),
            
            make_pipeline(FunctionTransformer(self.get_float_future, validate=False),StandardScaler())
            
#             make_pipeline(FunctionTransformer(self.get_binary_future, validate=False))
            
            
        ])
        
        return pipline.fit_transform(self.df)
        
```

```python
from sklearn.preprocessing import  LabelEncoder, LabelBinarizer, OneHotEncoder
from sklearn.pipeline import make_union, make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, Imputer
from sklearn.preprocessing import CategoricalEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import time

class Simple_pipline:
    
    '''
    Класс отвечает за преобразование данных
    Принимает таблицу с тренеровочной выборкой
    Преобразует важные признаки для модели обучения
    '''
    
    def __init__(self, df_ml, predict = False):
        
        '''
        При создание класса подгрузим таблицу для обучения из файла
        
        '''
        # Если нам нужен только предикт то обучаемся на всей выборке! Здесь нужно доработать
        if predict:
            
            self.df = pd.read_csv('simple_train_data.csv')
            
        else:
            
            self.df = df_ml
        
    def get_category(self,df):
        '''
        Функция возвращает категориальные признаки для обработки CategorialEncoding
        
        '''
        
        df['Углеводы на 100гр'] = (df['Углеводы'] / df['Масса(гр.)']) * 100
        
        # Признак генерит категории продуктов по содержанию углеводов! От низко содерж то высоко
        df['Содержание на 100'] = df['Углеводы на 100гр'].apply(lambda x: 1 if x <= 6 else 
                                                            2 if x <= 11 else 3 if x <=17 else
                                                           4 if x <=25 else 5 if x <=40 else 6)
        

        return df[['Прием пищи','Категория продукта','Оценка продукта','Год', 'Месяц', 'День',
                   'День_недели','Наименование продукта','Содержание на 100',
                   ]]
    
    def get_float_future(self,df):
        '''
        Функция возвращает числовые данные которые будут учавствовать в обучении
        Будет обрбаотка StandartScaller
        
        '''
        # Сюда я попробую подобавлять еще признаки по продуктам
        # Глюкоза 'Углеводы' 'Сумма углеводов'
        df['Сумма углеводов'] = df.groupby('ID_meal')['Углеводы'].transform('sum')
        
        summ_carbo = data['Углеводы'].sum()
        summ_insul = data['Инсулин под глюкозу'].sum()
        
        df['Инсулина на углеводы в среднем'] = df['Углеводы'] / (summ_carbo /summ_insul)
        
        
        return df[['Углеводы','Сумма углеводов','Масса(гр.)','Инсулина на углеводы в среднем','Вес(кг.)',
                  'Возраст(лет)','Стаж заболевания(лет)']]
    
    def get_binary_future(self,df):
        '''
        Функция добавляет бинарные признаки! Пол, Образовательное учереждение, Тренировка, Болезнь 
        
        '''
        
        return df[['Детский сад','Тренировка','Болезнь']]
    
    def pipline(self):
        '''
        Функция запускает обработчки pipline!
        
        '''
        
        pipline = make_union(*[
            
            make_pipeline(FunctionTransformer(self.get_category, validate=False),CategoricalEncoder(encoding='ordinal')),
            
            make_pipeline(FunctionTransformer(self.get_float_future, validate=False),StandardScaler())
            
#             make_pipeline(FunctionTransformer(self.get_binary_future, validate=False))
            
            
        ])
        
        return pipline.fit_transform(self.df)
        
```

```python
from sklearn.preprocessing import  LabelEncoder, LabelBinarizer, OneHotEncoder
from sklearn.pipeline import make_union, make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, Imputer
from sklearn.preprocessing import CategoricalEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import time

class Shugar_pipline:
    
    '''
    Класс отвечает за преобразование данных
    Принимает таблицу с тренеровочной выборкой
    Преобразует важные признаки для модели обучения
    '''
    
    def __init__(self, df_ml, predict = False):
        
        '''
        При создание класса подгрузим таблицу для обучения из файла
        
        '''
        # Если нам нужен только предикт то обучаемся на всей выборке! Здесь нужно доработать
        if predict:
            
            self.df = pd.read_csv('simple_train_data.csv')
            
        else:
            
            self.df = df_ml
        
    def get_category(self,df):
        '''
        Функция возвращает категориальные признаки для обработки CategorialEncoding
        
        '''
        
        
        return df[['Прием пищи','Год', 'Месяц', 'День','День_недели']]
    
    def get_float_future(self,df):
        '''
        Функция возвращает числовые данные которые будут учавствовать в обучении
        Будет обрбаотка StandartScaller
        
        '''
        df['Перемножение'] = df['Сумма углеводов'] * df['Сумма инсулина'] * df['Сахар перед едой']        
    
        # Сюда я попробую подобавлять еще признаки по продуктам
        # Глюкоза
        return df[['Сахар перед едой','Сумма углеводов',
                   'Сумма инсулина','Перемножение']]
    
    
    
    def pipline(self):
        '''
        Функция запускает обработчки pipline!
        
        '''
        
        pipline = make_union(*[
            
            make_pipeline(FunctionTransformer(self.get_category, validate=False),CategoricalEncoder(encoding='ordinal')),
            
            make_pipeline(FunctionTransformer(self.get_float_future, validate=False),StandardScaler())
            
            
        ])
        
        return pipline.fit_transform(self.df)
        
```
