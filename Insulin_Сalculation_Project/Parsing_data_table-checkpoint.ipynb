```python
import bs4 as bs
import urllib
import urllib.request
import re
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
def parsing_product_data(url):
    '''
   
    Функция парсит таблицу с продуктами с сайта и преобразует текщие данные в dataframe
    возвращает dataframe

    
    '''
    # Собираем все данные со страницы в 1объект
    urlpage = urllib.request.urlopen(url)
    soup = bs.BeautifulSoup(urlpage, 'html.parser')
    
    # Достаем данные которые нас интересуют из html
    
    core_list = []
    for data in soup.find_all('td'):
        core_list.append(data.text)
    
    # Чистим данные и достаем только те которые нам нужны
    clear_lists = []

    for index in range(len(core_list)):
        
        # Для отбора паттерн прост, после названия продукта идет число
        # Проверка на слово
        match = re.match(r"^[а-яА-ЯёЁ]", core_list[index])

        if match:

            try:
                # Если после слова ошибка не кидается то это наш слуай
                int(core_list[index + 1])
                
                # Генерируем списском продукт и его параметры! Попутно подчистим данные
                new_list = ['0' if core_list[element]=='\n—\n' else core_list[element] for element in  range(index,index + 7)]

                clear_lists.append(new_list)


            except:
                # Если ошибка вылетает то скидываем цикл и продолжаем итеации
                continue
                
    return  clear_lists, pd.DataFrame(clear_lists,columns=['Название','Гипоглекимический индекс','ККал',
                                                 'Белки','Жиры','Углеводы','Оценка продукта'])         
```

```python
# Думаю что мне нужно 2 функции 1 ищет ровно такой же при создание продукта!
# Вторая ищет похожие
def search_product_in_table(data_table,name_product):
    '''
    Функция открывает таблицу и ищит продукт по названию если он есть то выводит, если нет то пишит что продукта нет
    Можно добавлять
    
    '''
    for name in data_table['Наименование продукта'].values:
        
        if name_product.lower() == name.lower():
        
            print('Продукт есть в базе!')
            
            print(data_table[data_table['Наименование продукта'] == name].to_string())

            return True


    print('Продукта с таким именим нет в базе!')
    return False

    
    
    
```

## Итак теперь напишим функцию добавляющую новый продукт в таблицу

```python


def create_new_product(name_table, dict_new_row):
    '''
    Функция принимает параметры и создает новый датафрейм который можно конкатенировать с 
    
    текущей таблицей! Соединяет новый продукт с общей таблицей и пересохраняет таблицу
    
    Принимает: 
    
    1. Словарь введеных параметров(Последний параметр это категория продукта)
    2. Категория к которой относится продукт
    
    '''
    
    data_table = pd.read_csv(name_table)
    
    # Сначало проверим есть ли продукт в базе?
    
    name = dict_new_row['Наименование продукта']
    
    if search_product_in_table(data_table,name):
        
        print('Продукт с таким именем уже есть в базе!')
        
        return 
        
        
    else:    # продукта нет, можно создавать
        

        columns = data_table.columns


        data_row = pd.DataFrame(columns = columns,data = [dict_new_row])


        # Сделаем список по которому будем пробегатся и вставлять значения из расчета общей таблицы
        # В этом списке будут значения которые человек не ввел при добавление продукта

        none_columns = columns.difference(dict_new_row.keys())

        # уберем единственный интовый показатель который нужно расчитать отдельно 
        for col in none_columns.drop('Оценка продукта'):
            

            # Сгрупируй по категории по считай средние значения для пропущенного параметра
            data_row[col] = round(data_table[data_table['Категория продукта'] == dict_new_row['Категория продукта']][col].mean(),1)

            
            
                
        
            
    # Посчитаем медианную оценку по категории продукта и получим оценку           
    data_row['Оценка продукта'] = int(data_table[data_table['Категория продукта'] == dict_new_row['Категория продукта']]['Оценка продукта'].median())  
    
    # Обновляем продуктовую таблицу и перезаписываем ее
    data_table = data_table.append(data_row,ignore_index=True)
    
    data_table.to_csv(name_table,index=False)
    print('Продукт добавленн!Таблица перезаписанна!')
    
    print(data_row)    
    
    
```

```python
table_product_name = 'product_data.csv'

def create_new_product(dict_new_row, category):
    '''
    Функция принимает параметры и создает новый датафрейм который можно конкатенировать с 
    
    текущей таблицей! Соединяет новый продукт с общей таблицей и пересохраняет таблицу
    
    Принимает: 
    
    1. Словарь введеных параметров(Последний параметр это категория продукта)
    2. Категория к которой относится продукт
    
    
    '''
    data_table = pd.read_csv(table_product_name)
    
    columns = data_table.columns
    
    
    data_row = pd.DataFrame(columns = columns,data = [dict_new_row])
    
    
    # Сделаем список по которому будем пробегатся и вставлять значения из расчета общей таблицы
    # В этом списке будут значения которые только None
    
    none_columns = columns.difference(dict_new_row.keys())
    
    # уберем единственный интовый показатель который нужно расчитать отдельно 
    for col in none_columns.drop('Оценка продукта'):
        
        if data_table[col].dtype == 'float64':
            
            # Сгрупируй по категории по считай средние значения для пропущенного параметра
            data_row[col] = data_table[data_table[category] == 1].groupby(category)[col].mean().values[0]
        
            
        elif data_table[col].dtype == 'int64':  
            
            data_row[col] = data_row[col].fillna(0).astype('int64')
            
        else:
            continue
            
                
        
            
    # Посчитаем медианную оценку по категории продукта и получим оценку           
    data_row['Оценка продукта'] = data_table[data_table[category] == 1].groupby(category)['Оценка продукта'].median().values[0]   
    
    data_table.append(data_row,ignore_index=True).to_csv('table_product_name.csv',index=False)
    
    return data_row        
    
    
```
