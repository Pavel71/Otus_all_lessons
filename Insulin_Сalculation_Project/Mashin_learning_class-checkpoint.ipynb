```python
from sklearn.preprocessing import  LabelEncoder, LabelBinarizer, OneHotEncoder
from sklearn.pipeline import make_union, make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, Imputer
from sklearn.preprocessing import CategoricalEncoder
from sklearn.base import BaseEstimator, TransformerMixin
import datetime 
from datetime import timedelta
from dateutil import parser
import json
import pandas as pd

class Person_data:
    '''
    Класс создает новый датафрейм по заявке клиента на расчет инсулина
    Передает датафрейм который потом будет конкатенироватся с продуктовой таблицей!
    
    '''
    
    def __init__(self,new_id_meal, place_injection = 'Живот', human_param_json = 'param_dict.json',
                time_injection = None, time_meal = None, active = False, insulin_typ = True,
                temperatur_norm = True):
        
        '''
        Передадим все атрибуты которые потребуются для создания DataFrame
        
        
        '''
        
        '''
        
        Инициализатор принимает имя таблицы и подрубает DataFrame
        При создание класса мы передаем параметры которые человек должен сам указать,
        Это динмаичные параметры
        
    
        
        Принимает time_injection = str '10:15' - Время когда сделанн укол
        Принимает time_meal = str '10:30' - Время когда пища начинает поступать в организм
        Принимает булевое active - Планируется ли тренировка после приема пищи или нет?
        Принимает place_injection - Строка указывающая место инъекции
        Принимает insulin_typ - Тип инсулина Если True значит короткий под еду если нет Значит длинный
        Кто с посмпой у них всегда короткий! 
        Принимает temperatur_norm - Бинарный признак Температура тела внорме или повышенная(Болезнь)
        По умлочанию температура в норме
        count_product- Кол-во продуктов которое будем потреблять
        shugar_now- Текущий сахар - передается человеком
        '''
    
        
        self.time_injection = time_injection
        self.time_meal = time_meal
        self.active = active
        self.place_injection = place_injection
        self.insulin_typ = insulin_typ
        self.temperatur_norm = temperatur_norm
        self.new_id_meal = new_id_meal
        
        
        self.human_param_json = human_param_json
        
        # Считываем словарик
        with open(human_param_json,'r') as f:

            self.param_dict = json.load(f)
            
        # Создаем датафрейм куда запихнем статичные данные
        self.df = pd.DataFrame(index = range(1))
        
        # index = range(self.count_product)
            
        
        
        

    
    def time_future(self,df):
        
        '''
        Функция преобразует признак время на несколько других признаков!
        Время это больше статистика
        
        --------------------------------------------------- 
        Это к разряду что нужно будет подавать модели для
        обучения! А что идет просто для отображения статистики и визуализации
        
        А вот время приема пищи вполне может влиять на усвоение сахара!
        '''
       
        # Ставим текущее время    
        df['Время'] = datetime.datetime.now()
        
        
        df['Год'] = df['Время'].apply(lambda x: x.year)
        df['Месяц'] = df['Время'].apply(lambda x: x.month)
        df['День'] = df['Время'].apply(lambda x: x.day)
        df['День_недели'] = df['Время'].dt.weekday_name
        df['Детский сад'] = df['День_недели'].apply(lambda x: 0 if x == 'Saturday' or x == 'Sunday' else 1)
        
        # Добавим время приема пищи
        df['Прием пищи'] = df['Время'].dt.time.astype('str').apply(lambda x: 'Завтрак' if x < '10:00' else 
                                                                       'Обед' if x < '15:00' else 
                                                                       'Полдник' if x < '17:00' else 'Ужин')
    
        # Добавим Сезоны! времена года! Это может влиять на усваиваемось инсулина из за повышенной активности летом!
        # И частотой простуды зимой!
        df['Сезон'] = df['Месяц'].apply(lambda x: 'Весна' if 3 <= x < 6 else 
                                                      'Лето' if 6 <= x < 9 else 
                                                      'Осень' if 9 <= x <= 11 else 'Зима')
        
        return df[['Время','Год','Месяц','День','День_недели','Прием пищи','Сезон','Детский сад']]
    
    ###################################################################
    
    # Здесь идут статичные данные которые мы подгружаем из словарика! Они редко меняются!
    
    def get_human_param(self,df):
        
        '''
        Функция устанавливает все статичные параметры из словарика!
        Возвращает признаки в DataFrame
        '''
        
        # Создадим словарь с теми ключами которые достаточно просто записать! 
        #Остальные требуют отдельной обработки
        
        iter_dict = dict((x,self.param_dict[x]) for x in self.param_dict.keys() if x !='Инсулин под еду' 
                                                                                if x !='Инсулин для организма'
                                                                                if x != 'День рождения'
                                                                                if x != 'Дата диагноза')
        # Делаем сортировку и записываем в df
        for key,value in sorted(iter_dict.items()):
            
            df[key] = value
            
        return  df[sorted(iter_dict.keys())]  
    
    def age_diagnos_and_age(self, df):
        '''
        Функция принимает DataFrame

        Отсчитывает разницу со дня определения диагноза и по сегодняшний день!
        
        Берем Даты рождения и Дату диагноза из словарика!

        ''' 
        # Достаем строки из словаря и преобразуем их в datetime
        dt_diagnos = parser.parse(self.param_dict['Дата диагноза'],dayfirst=True)
        dt_birhtday = parser.parse(self.param_dict['День рождения'],dayfirst=True)
        
                                        # подсчитаем дельту  31536000 число секунд в году!         
        df['Стаж заболевания(лет)'] = df['Время'].apply(lambda x: (x -  dt_diagnos).total_seconds() // 31536000)

        df['Возраст(лет)'] = df['Время'].apply(lambda x: (x - dt_birhtday).total_seconds() // 31536000)

        # Проставим стаж заболевания и возраст в каждой строке в зависимости от времени
        return df[['Стаж заболевания(лет)','Возраст(лет)']]
    
    
    def time_injection_and_time_meal(self, df):
        '''
        Функция принимает DataFrame
        Принимает данные по времени введения инсулина и времени обеда

        Если передаются параметры time_injection str(10:00) и параметр time_meal str(10:15)
        то устанавливаем их! 
        Если нет то время инъекции берем текущее а время обеда + 10 мин

        '''

        # Преобразуем присланное время в datetime строку! Это нужно чтобы правельно прибавить 10минутный отступ за едой

        if self.time_injection is not None:

            df['Время инъекции'] = self.time_injection

        else:
            # Если время инъекции не переданно то считаем что инъекция делается в момент запроса!

            df['Время инъекции'] = df['Время'].apply(lambda x: x.strftime('%H:%M'))

        if self.time_meal is not None:

            df['Время приема пищи'] = self.time_meal

        else:
            # Если время приема пищи не переданные  то ставим + 10 минут от начала инъекции!
            df['Время приема пищи'] = df['Время'].apply(lambda x: (x + timedelta(minutes = 10)).strftime('%H:%M'))
        
        return  df[['Время приема пищи','Время инъекции']]
    
    
    def active_mod(self, df):
    
        '''
        Функция принимает DataFrame
        Если у человека планируется физическая нагрузка в течениее 1 часа после еды! То лучше это отобразить
        Так как актиный режим понижает уровень сахара в крови! Отсюда нужна корректировка дозировки!

        '''

        df['Тренировка'] = 1 if self.active else 0
        
        return df['Тренировка'].values.reshape(-1,1)
    
    def injection_place(self, df):
        '''
        Функция принимает DataFrame
        str(place) - Отображающий место укола Живот, Нога, Рука, Ягодицы
        Эта информация больше нужна для статитсики! Но есть научное подтверждение тому что с утра рекомендуется 
        делать укол в живот вечером же наоборот нельзя из за скорости усвоения инсулина человеком 
        в зависимости от времени суток! Это все касается при использование инсулиновых ручек или шприцов!
        С помпами этот параметр не так важен!
        '''
        
        # Здесь пропишу в основном для конретного объекта
        # Если утро то делаем укол в живот, в остальных случаях в руку
        
        now = datetime.datetime.now()
        
        if now.hour <= 10:
            
            df['Место инъекции'] = 'Живот'
            
        else:
            
            df['Место инъекции'] = 'Рука'
            
        
        return df['Место инъекции'].values.reshape(-1,1)
    
    def type_insulin(self, df):
        '''
        Функция принимает DataFrame
        Указывает тип инсулина -Короткий под еду Длинный для поддержания организма

        '''
        # Если True значит Укол под еду
        if self.insulin_typ:
            
            df['Тип инсулина'] = self.param_dict['Инсулин под еду']
        # Если нет значит делаем инъекцию длинного инслуина для организма    
        else:
            
            df['Тип инсулина'] = self.param_dict['Инсулин для организма']
            
        return df['Тип инсулина'].values.reshape(-1,1)  
    
    def temperature_body(self, df):
        '''
        Функция принимает DataFrame 
        Булевое значение norm = True по умолчанию! Значит температуры нет
        Если ставим False то температура!(Человек болеет)
        '''

        df['Болезнь'] = 0 if self.temperatur_norm else 1
        
        return df['Болезнь'].values.reshape(-1,1)
    
    def set_shugar_insulin(self, df):
        '''
        Функция просто устанавливает сахар в данный момент времени
        Добавляет 2 признака которые будут известны позже
        
        '''
        
        
        # Добавим признак сахар после еды(Он будет известен при следующем обращении)
        df['Сахар после еды'] = 0.0
        
        # Добавим признак инсулин! Его будем сетить после предикта
        
        df['Инсулин под глюкозу'] = 0
        
        df['Корректирующая доза'] = 0
        
        return df[['Сахар после еды','Инсулин под глюкозу','Корректирующая доза']]
    
    
        
    
    def set_id_meal(self, df):
        '''
        Функция устанавливает в DataFrame id_meal
        Добавляет +1 и перезаписывает в файл
        
        '''
        
        df['ID_meal'] = self.new_id_meal
        
            
        return df['ID_meal'].values.reshape(-1,1)   
            
   


    ######################################################################################    
    def get_pipline(self):
        '''
        Функция запускает пайплайн для преобразования данных из программы Freestyle_libre
        Дергает все методы постепенно
            
        '''  
        pipeline = make_union(*[
            
            
            make_pipeline(FunctionTransformer(self.time_future, validate=False)),
            
            make_pipeline(FunctionTransformer(self.get_human_param, validate=False)),
            
            make_pipeline(FunctionTransformer(self.type_insulin, validate=False)),
            
            make_pipeline(FunctionTransformer(self.age_diagnos_and_age, validate=False)),
            
            make_pipeline(FunctionTransformer(self.time_injection_and_time_meal, validate = False)),
            
            make_pipeline(FunctionTransformer(self.active_mod, validate = False)),
            
            make_pipeline(FunctionTransformer(self.injection_place, validate = False)),
           
            make_pipeline(FunctionTransformer(self.set_id_meal, validate = False)),
            
            make_pipeline(FunctionTransformer(self.temperature_body, validate = False)),
            
            make_pipeline(FunctionTransformer(self.set_shugar_insulin, validate = False))
                            
                                  ])          
            
        
        pipeline.fit_transform(self.df)
        
        # Пока буду возвращать готовый DataFrame Потом посмотрю!
        return self.df
```

```python
import functools
from difflib import get_close_matches

class Product_Table():
    
    '''
    Этот класс работает с табличкей с продуктами!
    
    Он может изменять значения в таблице!
    Добавлять новые продукты в таблицу!
    Выводит информацию о конкретных продуктах или целых категориях
    
    '''
    
    def __init__(self,name_table):
        
        '''
        Инициализируем при создание объекта класса DataFrame
        '''
        
#         self.new_id_meal = new_id_meal
        # Имя таблицы
        self.name_table = name_table
#         Откроем Табличку в DataFrame
        self.data_table = pd.read_csv(self.name_table, sep=',')
    
        
    
    
    # Запилим декоратор сохранения в csv таблицу
    
    def to_save_csv(func):
        '''
        Декоратор! Записывает DattaFrame в csv
        
        '''
        @functools.wraps(func)
        def wrapper(self, *args, **kwargs):
            
            result = func(self,*args, **kwargs)
            
            result.to_csv(self.name_table,index=False)
            
        return wrapper   
    
    @to_save_csv
    def add_new_product(self,dict_new_row):
        '''
        Функция принимает параметры и создает новый датафрейм который можно конкатенировать с 
        текущей таблицей! Соединяет новый продукт с общей таблицей и пересохраняет таблицу
        Принимает: 
        1. Словарь введеных параметров(Последний параметр это категория продукта)
        2. Категория к которой относится продукт
        '''


        # Сначало проверим есть ли продукт в базе?

        name = dict_new_row['Наименование продукта']

        if self.search_product_in_table(name):
            print('Продукт с таким именем уже есть в базе!')
            return 

        else:    # продукта нет, можно создавать
            columns = self.data_table.columns
            data_row = pd.DataFrame(columns = columns,data = [dict_new_row])
            
            # Сделаем список по которому будем пробегатся и вставлять значения из расчета общей таблицы
            # В этом списке будут значения которые человек не ввел при добавление продукта
            none_columns = columns.difference(dict_new_row.keys())
            
            # уберем единственный интовый показатель который нужно расчитать отдельно 
            for col in none_columns.drop('Оценка продукта'):

                # Сгрупируй по категории по считай средние значения для пропущенного параметра
                data_row[col] = round(self.data_table[self.data_table['Категория продукта'] == dict_new_row['Категория продукта']][col].mean(),1)

        # Посчитаем медианную оценку по категории продукта и получим оценку           
        data_row['Оценка продукта'] = int(self.data_table[self.data_table['Категория продукта'] == dict_new_row['Категория продукта']]['Оценка продукта'].median())  

        # Обновляем продуктовую таблицу и перезаписываем ее
        self.data_table = self.data_table.append(data_row,ignore_index=True)

       
        print('Продукт добавленн!Таблица перезаписанна!')

        print(data_row.to_string()) 
        
        return self.data_table
        
    def search_product_in_table(self, name_product, return_index_prod = False):
        '''
        Функция открывает таблицу и ищит продукт по названию если он есть то выводит, если нет то пишит что продукта нет
        Можно добавлять

        '''
        for name in self.data_table['Наименование продукта'].values:

            if name_product.lower() == name.lower():

                print('Продукт есть в базе!')

                print(self.data_table[self.data_table['Наименование продукта'] == name].to_string())

                # Если стоит задача вернуть и индекс продукта то она будет выполнена!
                if return_index_prod:

                    return self.data_table[self.data_table['Наименование продукта'] == name].index[0]

                return True


        print('Продукта с таким именим нет в базе!')
        return False    
    
    def find_product_data(self, name_product):
        '''
        Функция выведит данные по продукту
        Если продукта нет то напишит что нет

        '''
        
        
        if self.search_product_in_table(name_product):

            print('Удачного дня!')


        else: # продукта нет, можно создавать! Или есть продукт с входящими словами! Возможно нужен он!

            print('Пробую найти похожие названия!')
            # Применим стороннею библиотеку которая поможет найти похожие слова и фразы!
            
            list_name = self.data_table['Наименование продукта'].values
            
            # Вызовем специальный метод который проверяет слова на похожесть!
            
            list_close = get_close_matches(word = name_product, possibilities=list_name, n=3,cutoff=0.6)

            print('Возможно вам нужны эти названия: {}'.format(list_close))
                
#             for name in list_close:

# #                 # Беру слово без окончаний и понижаю во избежание ошибок
# #                 if name_product[:-1].lower() in name.lower():

# #                 display(self.data_table[self.data_table['Наименование продукта'] == name])
#                   print('Возможно вам нужны эти названия: {}'.format(name))
    
    @to_save_csv
    def delete_product(self,name_product):
        '''
        Функция принимает имя продукта! И удаляет его из таблицы
        после обновляет таблицу!
        
        '''
        
        self.data_table.drop(self.data_table[self.data_table['Наименование продукта'] == name_product].index[0],
                                                                                         inplace = True)
        
        print('Продукт удаленн!')
        
        return self.data_table
        
      
        
        
    def find_index_product(self, name_product):
        '''
        Функция ищет объект по имени и выводит его индекс!

        '''

        # Если продукт есть напичатает что есть и вернет индекс этого продукта
        
        index = self.search_product_in_table(name_product, return_index_prod = True) 
        
        print('Индекс данного продукта {}.'.format(index))

        return index  
    
    def print_category(self):
        '''
        Функция выводит на экран все доступные категории продуктов
        
        '''

        display(self.data_table.groupby('Категория продукта').size().sort_values(ascending= False))
        
    def print_product_on_category(self,category):
        '''
        Функция выводит все продукты по конкретной категории
        
        '''
        display(self.data_table[self.data_table['Категория продукта'] == category]) 
       
    @to_save_csv
    def update_product_table(self, identificator, dict_change):
        '''
        Функция изменяет любой параметр в объекте!
        Принимает Идентификатор - это либо индекс либо имя продукта
        dict_change - словарь с значениями которые мы перезаписываем
        
        '''

        # Если Если старший класс инт Нам передали индекс
        if isinstance(identificator,int):
            
            for name_future in dict_change:
                
                self.data_table.at[identificator, name_future] = dict_change[name_future]
                
            print('Данные обновленны! Записываю в таблицу')
                
            display(self.data_table.loc[identificator])
            
        else:
            
            # Если попали сюда значит нам передали название продукта
            
            index = self.find_index_product(identificator)
            
            # Дальше код просто копируется
            
            for name_future in dict_change:
                
                self.data_table.at[index, name_future] = dict_change[name_future]
                
            print('Данные обновленны! Записываю в таблицу')
                
            display(self.data_table.loc[index])
            
        return self.data_table
            
            
    def  get_product(self, name_product: str, new_mass: int, new_id_meal: int):
        '''
        Функция принимает имя продукта и его массу
        Возвращает DataFrame 1ой строкой(Посмотрю может проще словарем)
        name_product - название продукта
        new_mass- масса продукта
        id_meal-идентификатор приема пищи -
        '''
        
        # Добавим проверку на вхождение имени продукта в таблицу продуктов
        
        
        if name_product in self.data_table['Наименование продукта'].values:
            
            new_data = self.data_table[self.data_table['Наименование продукта'] == name_product].reset_index().drop('index',axis=1)

            for col in ['Ккал','Белки','Жиры','Углеводы','Глюкоза']:


                new_data[col] = new_data[col] * new_mass / new_data['Масса(гр.)']

            new_data.loc[0,'Масса(гр.)'] = new_mass

            new_data['Глюкоза'] = round(new_data['Гликемический индекс'] * new_data['Углеводы'] /100,2)

            # Установим id_meal по нему будем joinить
            new_data.loc[0,'ID_meal'] = new_id_meal

            # Все здесь мы вернули данные По продукту в зависимости от массы!
            return new_data
        
        else:
            
            print('Имя продукта "{}" не найдено! Проверьте правильность или найдите похожие!'.format(name_product))
            
            return False
    
        
        
```

```python
def correct_insulin(data, meal_id, shugar):
    '''
    Функция принимает
    
    Функция анализирует насколько была хорошей компенсация предыдущего обеда!
    Хорошей компенсацией считается сахар после обеда от 70 до 145
    И исходя из текущего сахара вносит поправку в дозировку для предыдущего обеда!
    
    
    '''
    
    # Определим как считать поправку! На каждые 10п сахара выше или ниже нормы нужно
    # Нужно добавить или убрать 0.1 еденицу инсулина
    
    # Используем тернарный оператор
    correct_insulin = (-(70 - shugar) * 0.01) if shugar < 70 else ((shugar - 150) * 0.01) if shugar > 150 else 0
    
    if correct_insulin != 0:
        
        # Здесь по идеи нужно написать подфункцию! Что бы не дублировать кода!
    
        df = data.loc[data['ID_meal'] == meal_id]


        # Сумма углеводов под обед
        sum_carbo =  df['Сумма углеводов'].values[0]

        df['Поправка инсулина'] = df['Углеводы'].apply(lambda x: x / sum_carbo * correct_insulin )

        df['Новый инсулин под глюкозу'] = df['Инсулин под глюкозу'] + df['Поправка инсулина']
        
        return df['Новый инсулин под глюкозу'].values
    
    else:
                # Если корректировки нет то возвращаем тоже самое
        return  df['Инсулин под глюкозу'].values
    
    
    # Все вернули список новых значений и подставили в предыдущий обед! На этом функция корректировки завершена
    # Теперь следует все это связать в 1 программу и запилить проект в pycharm
    
```

```python
from sklearn.preprocessing import  LabelEncoder, LabelBinarizer, OneHotEncoder
from sklearn.pipeline import make_union, make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, Imputer
from sklearn.preprocessing import CategoricalEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import time


class Mashin_learning():
    
    '''
    Этот класс 
    
    1. Сделает предсказание корректирующего коэффициента
    2. Подготовит нужные признаки и сделает предсказание инсулина для каждого продукта
    3. Подсчитает общую сумму инсулина с учетом корректирующего коэффициента
    4. Сделает предсказание сахара после еды на основе полученных данных!
    5. Заполнит таблицу корректно и вернет нужные ответы
    
    
    '''
    
    def __init__(self, train_data, count_prod):
        
        '''
        
        train_data = DataFRame c данными для обучения и предсказания
        count_prod = Кол-во продуктов
        
        
        '''
        
        self.train_data = train_data
        self.count_prod = count_prod
        
    def work_pred_meal(self,test_index, last_index, meal_id ,pred_id_meal):
        '''
        Функция отвечает за работу с предыдущим обедом
        Мы ставим текущий сахар в колонку сахар псоле еды за предыдущий обед
        Если текущий сахар высокий то поправляем дозировку инсулина
        
        С другой стороны это помешает делать предсказания сахара! т.к как модель будет тренироватся на сахаре
        который не соответвует действительности
        Нужно по всей видимости вводить новый признак инсулин под глюкозу после поправки
        и обучатся надо на нем! А делать предсказания сахара на обычном инсулине
        и если поправлять не надо то будет все норм а если надо
        
        
        '''
        
        
        # Берем текущий сахар
        shugar_now = self.train_data.loc[last_index,'Сахар перед едой']
        
        
        # Ставим в предыдущий обед сахар который получили после компенсациии предыдущего обеда!
        self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal, 'Сахар после еды'] = shugar_now
        
        # Если сахар сейчас выше или ниже нормы! ТО следует изменить дозировку на предыдущем обеде
        # Что бы моделька обучалась на обновленных данных
        
        df = self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal]
        
        # Для предыдущего обеда получис дозировку! Соглсано компенсации! Если компенсация успешная, то доз не меняем.
        cor_insul_shugar, corr_koef = self.correct_insulin_of_shugar(df=df,shugar=shugar_now)
        
        # Внесем изменения в DataFrame
        self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal,'Инсулин под глюкозу'] = cor_insul_shugar
        
        self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal,'Коэфициент изменения'] = corr_koef
        
        return cor_insul_shugar, corr_koef, shugar_now
        
        
        
        # Установим значение инсулина за предыдущий обед!
        # Опять же если сахар в норме после обеда то они не изменятся!
        

        
        
        
        
    def get_predict(self, test_index, last_index, meal_id):
        '''
        Функция вызывает методы и получает предсказания сахар, корр коэфициента от сахара
        Предсказание сахара + она сетит эти значения в DataFrame
        
        '''
        
        # Получили предсказание корректирующего коэффициента
        pred_koef = self.get_pred_korr_koef()
        
         
        # Получил предсказание инсулина под еду
        pred_insulin = self.get_pred_insulin_koef()
        
        print(f'Предсказанный сахар для текущего обеда {pred_insulin}')
        
        
        # Теперь стоит задача подсчитать сумму инсулина за обед с учетом корреткирующего коэф.
        
        # Поставим полученные значения
        
        
        self.train_data.loc[test_index:,'Корректирующая доза'] = pred_koef
        
        self.train_data.loc[test_index:,'Инсулин под глюкозу'] = pred_insulin
        
        self.train_data['Сумма инсулина'] = self.train_data.groupby('ID_meal'
                                                                   )['Инсулин под глюкозу'].transform('sum'
                                                                   ) + self.train_data['Корректирующая доза']
        
        # Все это можно возвращать! То что нам нужно!
        insulin_for_meal = self.train_data.loc[last_index,'Сумма инсулина']
        
        print(f'Сумма инсулина {insulin_for_meal}')
        
        
        
        # Подсчитаем Ожидаемый сахар от этой дозы инсулина и обеда!
        pred_shugar = self.get_pred_shugar_after_meal()
        
        self.train_data.loc[self.train_data['ID_meal'] == meal_id, 'Предсказанный сахар'] = pred_shugar
        
        
        
        
        return insulin_for_meal, pred_shugar
    
    def work_this_male(self, meal_id):
        '''
        Функция работает с текущим обедом! В случае если мы не согласны с дозировкой инсулина
        Мы попавляем значения над каждым продуктом в зависимости от разницы между предсказанной дозой и нашей
        
        
        '''
        
        accept_insul = input('Вы согласны с такой дозировкой? Введите ДА или свою дозировку на обед')
        
        
        ###  Если с дозой мы не согласны то перезаписываем значения
        
        try: 
            
            accept_insul = float(accept_insul)
        
            # Так как мы изменяем инсулин сейчас и еще не делали предсказание! то нам эта поправка особо не нужна
            # Мы прежде чем делать предсказание сахара поправим инсулин руками!
            
            
            corr_insul_human_check, _ = self.correct_insulin_of_human_check(meal_id, accept_insul)


            # Вносим изменения в текущий обед! теперь можно предсказывать сахар
            self.train_data.loc[self.train_data['ID_meal'] == meal_id, 'Инсулин под глюкозу'] = corr_insul_human_check
             
            # Для измененых данных нужно переситать сумму инсулина! Она по сути равна той которую человек ставит
            # Тогда прогноз сахара изменится
            self.train_data.loc[self.train_data['ID_meal'] == meal_id,'Сумма инсулина'] = accept_insul
            
            # Пересчитаем ожидаемый сахар в зависимости от новой дозы инсулина на обед
            pred_shugar = self.get_pred_shugar_after_meal()

            # Установи новый признак! предсказаннй сахар

            self.train_data.loc[self.train_data['ID_meal'] == meal_id, 'Предсказанный сахар'] = pred_shugar

            print('Ваш обед')

            meal = self.train_data.loc[self.train_data['ID_meal'] == meal_id,['ID_meal','Наименование продукта','Масса(гр.)','Инсулин под глюкозу']]

            print(meal)


            print('Вы выбрали {:.3} инсулина,Новый прогноз на сахар {:.4}({:.3}))'.format(accept_insul,pred_shugar,pred_shugar / 18))
        
        except ValueError:
            
            print('Ввели строку! Значит ничего не меняем!')
        
        
    def get_work(self,name_train_table):
        '''
        Функция запускает работу Алгоритма!
        Класс работает с таблицей, где содержится новые данные! Тоесть обед уже добавленн!
        Возвращает нужный результат! Преобразует таблицу и сохраняет ее!
        
        
        '''
        
        # Узнаем точно индексы где находятся новые данные
        last_index = self.train_data.index.values[-1]
        test_index = last_index - self.count_prod + 1
        
        meal_id = np.unique(self.train_data['ID_meal'].values)[-1]
        
        pred_id_meal = meal_id - 1
        
        
        
        ######################## Большой блок работы с предыдущем обедом ###########################
        
        
        
        # Если надо то подправим дозировку инсулина под предыдущим обедом + добавим сахар после еды
        cor_insul_shugar, corr_koef, shugar_now = self.work_pred_meal(test_index, last_index, meal_id ,pred_id_meal)
        
        
        print(f'Скорректированный инсулин за предыдущий обед!{cor_insul_shugar}, коэффициент от сахара сейчас {corr_koef}')
        
        # Если компенсация хорошая то не меняем! Если плохая то корректируем дозировку для обучения!
        
        

        
        ############################### Получение предсказаний на обучающейся выборке ###################
        
        insulin_for_meal, pred_shugar = self.get_predict(test_index, last_index, meal_id)
        
        
        
        meal = self.train_data.loc[self.train_data['ID_meal'] == meal_id,['ID_meal','Наименование продукта','Масса(гр.)','Инсулин под глюкозу']]
        
        
        print('На ваш обед')
        
        print(meal)
        
        print('Рекомендуется ввести {:.3} инсулина, ожидаемый сахар {:.4}({:.3}))'.format(insulin_for_meal,pred_shugar,pred_shugar / 18))
        
        
        ########################### Если не согласны с дозировкой то идет работа с текущим обедом ###############
        
        
        self.work_this_male(meal_id)
            

            

#         self.train_data.to_csv(name_train_table,index=False)
        
    def correct_insulin_of_human_check(self, meal_id, accept_insul):
        
        '''
        Функция корректирует дозировку инсулина, если человек не согласен с прогнозом алгоритма!
        Она больше нужна для отладки программы!
        Функция принимает corr_insul - Доза которую решает ввести человек лично
        Исходя из нее мы узнаем разницу между дозой алгоритма и человека и на эту разницу изменяем
        показатели Инсулин под еду для каждого продукта
        
        '''
        
        
        df = self.train_data.loc[self.train_data['ID_meal'] == meal_id]
            
        sum_insul = df['Сумма инсулина'].values[0]
            
        correct_insulin = accept_insul - sum_insul
            
        print('Внесли поправку! Передаю новые значениея')
            
        list_new_ins, corr_koef_this_meal = self.pod_correct_insulin(df, correct_insulin)
            
        # Если строка то возвращаем те же значения что и предсказали! Значит согласны
            
        return list_new_ins, corr_koef_this_meal
        
    def get_pred_shugar_after_meal(self):
        '''
        Функция которая делает предсказание сахара который мы ожидаем получить с этой компенсацией!
        
        '''
        # Получаем обработанные данные
        
        print('Обращаюсь к методу 2 раза! Оба разные',self.train_data.loc[self.train_data['ID_meal'] == 105, 'Инсулин под глюкозу'])
        
        
        train_pipline = Shugar_pipline(train_data = self.train_data).get_pipline()
        
        # Настройки были вычеслинны с помощью gridsearch
        xgb_shugar = XGBRegressor(max_depth=2,learning_rate=0.01,n_estimators=200)
        
        
        # Для обучения берем все кроме последних!
        xgb_shugar.fit(train_pipline[:-self.count_prod], self.train_data['Сахар после еды'][:-self.count_prod])

        pred = xgb_shugar.predict(train_pipline[-self.count_prod:])
        
        return pred[0]
        
        
        
        
    def get_pred_korr_koef(self):
        
        '''
        Функция возмет таблицу и сделает предсказание корректирующего коэффициента
        
        Этот коэффициент отвечает за + или - к общей дозе инсулина в зависимости
        от сахара до еды! Скажем если сахар выше норм околок 180 то нужно скорректировать его 
        дополнительной дозой инсулина в размере 0.5
        А если ниже например 40 - то нужно наоборот сделать инъекцию меньше на 0.5
        
        '''
        # Так как новые данные добавляются в конец таблицы
        train_koef = self.train_data['Сахар перед едой'][:-self.count_prod].values.reshape(-1,1)
        
        target_koef = self.train_data['Корректирующая доза'][:-self.count_prod].values.reshape(-1,1)
        
        test_koef = self.train_data['Сахар перед едой'][-self.count_prod:].values.reshape(-1,1)

        
        xgb_koef = XGBRegressor(learning_rate=0.1,max_depth=2,n_estimators=50)
        
        xgb_koef.fit(train_koef,target_koef)
        
        predict = xgb_koef.predict(test_koef).round(decimals=2)
        
        return predict
    
    def get_pred_insulin_koef(self):
        
        '''
        Функция Делает предсказание дозы инсулина под каждый продукт
        Возвращает дозировки
        
        '''
        # Получим преобразованный DataFrame!
        train_pipline = Simple_pipline(train_data = self.train_data).get_pipline()
        
        xgb_insulin = XGBRegressor(max_depth=2,learning_rate=0.05,n_estimators=150)
        
        # Для обучения берем все кроме последних!
        xgb_insulin.fit(train_pipline[:-self.count_prod], self.train_data['Инсулин под глюкозу'][:-self.count_prod])

        pred = np.round(xgb_insulin.predict(train_pipline[-self.count_prod:]),decimals =2)
        
        return pred
    
    def correct_insulin_of_shugar(self, df, shugar):
        '''
        Функция изменяет дозировку инсулина за предыдущий обед если мы ошиблись! Ошибка зависит от отклонения 
        сахара от нормы

        Функция анализирует насколько была хорошей компенсация предыдущего обеда!
        Хорошей компенсацией считается сахар после обеда от 70 до 145
        И исходя из текущего сахара вносит поправку в дозировку для предыдущего обеда!

        '''

        # Определим как считать поправку! На каждые 10п сахара выше или ниже нормы нужно
        # Нужно добавить или убрать 0.1 еденицу инсулина
           
            
            # Используем тернарный оператор
        correct_insulin = ((shugar - 70) * 0.01) if shugar < 70 else ((shugar - 150) * 0.01) if shugar > 150 else 0
        
        
        if correct_insulin != 0:
            
            print(f"Сахар вне нормы требуется внести корректировку! {correct_insulin}")
            

            return self.pod_correct_insulin(df,correct_insulin)

        else:
                        # Если корректировки нет то возвращаем тоже самое
            
            print('Компенсация обеда прошла успешно!')
                    
                    # Если сахар в норме то возвразщаем обычные значения
            return  df['Инсулин под глюкозу'].values, correct_insulin
        
        
    def pod_correct_insulin(self, df, correct_insulin):
            
            '''
            Функция содежит часть кода от большой функции! так она нам понадобится в другом месте программы
            Принимает DataFrame - df
            Принимает correct_insulin - (Пример 0.3) Коэффициент на который нам нужно изменить сумму инсулина
            и соответвенно все дозировки по отдельным продуктам
            
            Например у нас 2 продукта
            
            Апельсин 100г Углеводы 10г - Инсулин под глюкозу 1.0
            Виноград 100г Углеводы 15г - Инсулин под глюкозу 1.5
            
            Скажем алгоритм выдал предсказание сделать 2.5 ед инсулина! Сахар после этой компенсации стал = 200
            
            Выше нормы на 60! Мы вычеслаем корректировку correct_insulin - 0.6 И нам нужно 0.6 инсулина добавить
            к прошлой компенсации! Распределяем мы 0.6 в пропорции от массы углеводов! Винограду достанется 0.4
            а апельсину 0.2 Прибавка к дозе!
            
            Станет:
            
            Апельсин 100г Углеводы 10г - Инсулин под глюкозу 1.2
            Виноград 100г Углеводы 15г - Инсулин под глюкозу 1.9
            
            
            Возвращает список новых значений.
            
            '''
            
            sum_carbo =  df['Сумма углеводов'].values[0]

            df['Поправка инсулина'] = df['Углеводы'].apply(lambda x: x / sum_carbo * correct_insulin)

            # Таким образом если сахар выше нормы то за предыдущий обед мы повысим дозу инсулина!
            # Если ниже то понизим
            df['Новый инсулин под глюкозу'] = df['Инсулин под глюкозу'] + df['Поправка инсулина']
            
            # Вернем новый инсулин и сам коэффициент корректировки
            return df['Новый инсулин под глюкозу'].values, correct_insulin
    
    
    
    

        
```

```python
from sklearn.preprocessing import  LabelEncoder, LabelBinarizer, OneHotEncoder
from sklearn.pipeline import make_union, make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, Imputer
from sklearn.preprocessing import CategoricalEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
import time


class Mashin_learning():
    
    '''
    Этот класс 
    
    1. Сделает предсказание корректирующего коэффициента
    2. Подготовит нужные признаки и сделает предсказание инсулина для каждого продукта
    3. Подсчитает общую сумму инсулина с учетом корректирующего коэффициента
    4. Сделает предсказание сахара после еды на основе полученных данных!
    5. Заполнит таблицу корректно и вернет нужные ответы
    
    
    '''
    
    def __init__(self, train_data, count_prod):
        
        '''
        
        train_data = DataFRame c данными для обучения и предсказания
        count_prod = Кол-во продуктов
        
        
        '''
        
        self.train_data = train_data
        self.count_prod = count_prod
        
    def work_pred_meal(self,test_index, last_index, meal_id ,pred_id_meal):
        '''
        Функция отвечает за работу с предыдущим обедом
        Мы ставим текущий сахар в колонку сахар псоле еды за предыдущий обед
        Если текущий сахар высокий то поправляем дозировку инсулина
        
        С другой стороны это помешает делать предсказания сахара! т.к как модель будет тренироватся на сахаре
        который не соответвует действительности
        Нужно по всей видимости вводить новый признак инсулин под глюкозу после поправки
        и обучатся надо на нем! А делать предсказания сахара на обычном инсулине
        и если поправлять не надо то будет все норм а если надо
        
        
        '''
        
        
        # Берем текущий сахар
        shugar_now = self.train_data.loc[last_index,'Сахар перед едой']
        
        
        # Ставим в предыдущий обед сахар который получили после компенсациии предыдущего обеда!
        self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal, 'Сахар после еды'] = shugar_now
        
        # Если сахар сейчас выше или ниже нормы! ТО следует изменить дозировку на предыдущем обеде
        # Что бы моделька обучалась на обновленных данных
        
        df = self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal]
        
        # Для предыдущего обеда получис дозировку! Соглсано компенсации! Если компенсация успешная, то доз не меняем.
        cor_insul_shugar, corr_koef = self.correct_insulin_of_shugar(df=df,shugar=shugar_now)
        
        return cor_insul_shugar, corr_koef, shugar_now
        
        
        
        # Установим значение инсулина за предыдущий обед!
        # Опять же если сахар в норме после обеда то они не изменятся!
        
#         self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal,'Инсулин под глюкозу'] = cor_insul_shugar
        
#         self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal,'Коэфициент изменения'] = corr_koef
        
        
        
        
    def get_predict(self, test_index, last_index, meal_id):
        '''
        Функция вызывает методы и получает предсказания сахар, корр коэфициента от сахара
        Предсказание сахара + она сетит эти значения в DataFrame
        
        '''
        
        # Получили предсказание корректирующего коэффициента
        pred_koef = self.get_pred_korr_koef()
        
         
        # Получил предсказание инсулина под еду
        pred_insulin = self.get_pred_insulin_koef()
        
        print(f'Предсказанный сахар для текущего обеда {pred_insulin}')
        
        
        # Теперь стоит задача подсчитать сумму инсулина за обед с учетом корреткирующего коэф.
        
        # Поставим полученные значения
        
        
        self.train_data.loc[test_index:,'Корректирующая доза'] = pred_koef
        
        self.train_data.loc[test_index:,'Инсулин под глюкозу'] = pred_insulin
        
        self.train_data['Сумма инсулина'] = self.train_data.groupby('ID_meal'
                                                                   )['Инсулин под глюкозу'].transform('sum'
                                                                   ) + self.train_data['Корректирующая доза']
        
        # Все это можно возвращать! То что нам нужно!
        insulin_for_meal = self.train_data.loc[last_index,'Сумма инсулина']
        
        print(f'Сумма инсулина {insulin_for_meal}')
        
        
        # Подсчитаем Ожидаемый сахар от этой дозы инсулина и обеда!
        pred_shugar = self.get_pred_shugar_after_meal(data = self.train_data)
        
        self.train_data.loc[self.train_data['ID_meal'] == meal_id, 'Предсказанный сахар'] = pred_shugar
        
        
        
        
        return insulin_for_meal, pred_shugar
    
    def work_this_male(self, meal_id):
        '''
        Функция работает с текущим обедом! В случае если мы не согласны с дозировкой инсулина
        Мы попавляем значения над каждым продуктом в зависимости от разницы между предсказанной дозой и нашей
        
        
        '''
        
        accept_insul = input('Вы согласны с такой дозировкой? Введите ДА или свою дозировку на обед')
        
        
        ###  Если с дозой мы не согласны то перезаписываем значения
        
        try: 
            
            accept_insul = float(accept_insul)
        
            # Так как мы изменяем инсулин сейчас и еще не делали предсказание! то нам эта поправка особо не нужна
            # Мы прежде чем делать предсказание сахара поправим инсулин руками!
            
            
            corr_insul_human_check, _ = self.correct_insulin_of_human_check(self.train_data, meal_id, accept_insul)


            # Вносим изменения в текущий обед! теперь можно предсказывать сахар
            self.train_data.loc[self.train_data['ID_meal'] == meal_id, 'Инсулин под глюкозу'] = corr_insul_human_check

            # Пересчитаем ожидаемый сахар в зависимости от новой дозы инсулина на обед
            pred_shugar = self.get_pred_shugar_after_meal(data = self.train_data)

            # Установи новый признак! предсказаннй сахар

            self.train_data.loc[self.train_data['ID_meal'] == meal_id, 'Предсказанный сахар'] = pred_shugar

            print('Ваш обед')

            meal = self.train_data.loc[self.train_data['ID_meal'] == meal_id,['ID_meal','Наименование продукта','Масса(гр.)','Инсулин под глюкозу']]

            print(meal)


            print('Вы выбрали {:.3} инсулина,Новый прогноз на сахар {:.4}({:.3}))'.format(accept_insul,pred_shugar,pred_shugar / 18))
        
        except ValueError:
            
            print('Ввели строку! Значит ничего не меняем!')
        
        
    def get_work(self,name_train_table):
        '''
        Функция запускает работу Алгоритма!
        Класс работает с таблицей, где содержится новые данные! Тоесть обед уже добавленн!
        Возвращает нужный результат! Преобразует таблицу и сохраняет ее!
        
        
        '''
        
        # Узнаем точно индексы где находятся новые данные
        last_index = self.train_data.index.values[-1]
        test_index = last_index - self.count_prod + 1
        
        meal_id = np.unique(self.train_data['ID_meal'].values)[-1]
        
        pred_id_meal = meal_id - 1
        
        data = self.train_data
        
        
        ######################## Большой блок работы с предыдущем обедом ###########################
        
        
        
        # Если надо то подправим дозировку инсулина под предыдущим обедом + добавим сахар после еды
        cor_insul_shugar, corr_koef, shugar_now = self.work_pred_meal(test_index, last_index, meal_id ,pred_id_meal)
        
        
        print(f'Скорректированный инсулин за предыдущий обед!{cor_insul_shugar}, коэффициент от сахара сейчас {corr_koef}')
        
        # Если компенсация хорошая то не меняем! Если плохая то корректируем дозировку для обучения!
        
        self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal,'Инсулин под глюкозу'] = cor_insul_shugar
        
        self.train_data.loc[self.train_data['ID_meal'] == pred_id_meal,'Коэфициент изменения'] = corr_koef
        

        
        ############################### Получение предсказаний на обучающейся выборке ###################
        
        insulin_for_meal, pred_shugar = self.get_predict(test_index, last_index, meal_id)
        
        
        
        meal = self.train_data.loc[self.train_data['ID_meal'] == meal_id,['ID_meal','Наименование продукта','Масса(гр.)','Инсулин под глюкозу']]
        
        
        print('На ваш обед')
        
        print(meal)
        
        print('Рекомендуется ввести {:.3} инсулина, ожидаемый сахар {:.4}({:.3}))'.format(insulin_for_meal,pred_shugar,pred_shugar / 18))
        
        
        ########################### Если не согласны с дозировкой то идет работа с текущим обедом ###############
        
        
        self.work_this_male(meal_id)
            

            

        self.train_data.to_csv(name_train_table,index=False)
        
    def correct_insulin_of_human_check(self, data, meal_id, accept_insul):
        
        '''
        Функция корректирует дозировку инсулина, если человек не согласен с прогнозом алгоритма!
        Она больше нужна для отладки программы!
        Функция принимает corr_insul - Доза которую решает ввести человек лично
        Исходя из нее мы узнаем разницу между дозой алгоритма и человека и на эту разницу изменяем
        показатели Инсулин под еду для каждого продукта
        
        '''
        
        
        df = data.loc[data['ID_meal'] == meal_id]
            
        sum_insul = df['Сумма инсулина'].values[0]
            
        correct_insulin = accept_insul - sum_insul
            
        print('Внесли поправку! Передаю новые значениея')
            
        list_new_ins, _ = self.pod_correct_insulin(df, correct_insulin)
            
        # Если строка то возвращаем те же значения что и предсказали! Значит согласны
            
        return list_new_ins, _
        
    def get_pred_shugar_after_meal(self, data):
        '''
        Функция которая делает предсказание сахара который мы ожидаем получить с этой компенсацией!
        
        '''
        # Получаем обработанные данные
        
        
        train_pipline = Shugar_pipline(train_data = data).get_pipline()
        
        # Настройки были вычеслинны с помощью gridsearch
        xgb_shugar = XGBRegressor(max_depth=2,learning_rate=0.01,n_estimators=200)
        
        
        # Для обучения берем все кроме последних!
        xgb_shugar.fit(train_pipline[:-self.count_prod], self.train_data['Сахар после еды'][:-self.count_prod])

        pred = xgb_shugar.predict(train_pipline[-self.count_prod:])
        
        print(pred)
        
        return pred[0]
        
        
        
        
    def get_pred_korr_koef(self):
        
        '''
        Функция возмет таблицу и сделает предсказание корректирующего коэффициента
        
        Этот коэффициент отвечает за + или - к общей дозе инсулина в зависимости
        от сахара до еды! Скажем если сахар выше норм околок 180 то нужно скорректировать его 
        дополнительной дозой инсулина в размере 0.5
        А если ниже например 40 - то нужно наоборот сделать инъекцию меньше на 0.5
        
        '''
        # Так как новые данные добавляются в конец таблицы
        train_koef = self.train_data['Сахар перед едой'][:-self.count_prod].values.reshape(-1,1)
        
        target_koef = self.train_data['Корректирующая доза'][:-self.count_prod].values.reshape(-1,1)
        
        test_koef = self.train_data['Сахар перед едой'][-self.count_prod:].values.reshape(-1,1)

        
        xgb_koef = XGBRegressor(learning_rate=0.1,max_depth=2,n_estimators=50)
        
        xgb_koef.fit(train_koef,target_koef)
        
        predict = xgb_koef.predict(test_koef).round(decimals=2)
        
        return predict
    
    def get_pred_insulin_koef(self):
        
        '''
        Функция Делает предсказание дозы инсулина под каждый продукт
        Возвращает дозировки
        
        '''
        # Получим преобразованный DataFrame!
        train_pipline = Simple_pipline(train_data = self.train_data).get_pipline()
        
        xgb_insulin = XGBRegressor(max_depth=2,learning_rate=0.05,n_estimators=150)
        
        # Для обучения берем все кроме последних!
        xgb_insulin.fit(train_pipline[:-self.count_prod], self.train_data['Инсулин под глюкозу'][:-self.count_prod])

        pred = np.round(xgb_insulin.predict(train_pipline[-self.count_prod:]),decimals =2)
        
        return pred
    
    def correct_insulin_of_shugar(self, df, shugar):
        '''
        Функция изменяет дозировку инсулина за предыдущий обед если мы ошиблись! Ошибка зависит от отклонения 
        сахара от нормы

        Функция анализирует насколько была хорошей компенсация предыдущего обеда!
        Хорошей компенсацией считается сахар после обеда от 70 до 145
        И исходя из текущего сахара вносит поправку в дозировку для предыдущего обеда!

        '''

        # Определим как считать поправку! На каждые 10п сахара выше или ниже нормы нужно
        # Нужно добавить или убрать 0.1 еденицу инсулина
           
            
            # Используем тернарный оператор
        correct_insulin = ((shugar - 70) * 0.01) if shugar < 70 else ((shugar - 150) * 0.01) if shugar > 150 else 0
        
        
        if correct_insulin != 0:
            
            print(f"Сахар вне нормы требуется внести корректировку! {correct_insulin}")
            

            return self.pod_correct_insulin(df,correct_insulin)

        else:
                        # Если корректировки нет то возвращаем тоже самое
            
            print('Компенсация обеда прошла успешно!')
                    
                    # Если сахар в норме то возвразщаем обычные значения
            return  df['Инсулин под глюкозу'].values, correct_insulin
        
        
    def pod_correct_insulin(self, df, correct_insulin):
            
            '''
            Функция содежит часть кода от большой функции! так она нам понадобится в другом месте программы
            Принимает DataFrame - df
            Принимает correct_insulin - (Пример 0.3) Коэффициент на который нам нужно изменить сумму инсулина
            и соответвенно все дозировки по отдельным продуктам
            
            Например у нас 2 продукта
            
            Апельсин 100г Углеводы 10г - Инсулин под глюкозу 1.0
            Виноград 100г Углеводы 15г - Инсулин под глюкозу 1.5
            
            Скажем алгоритм выдал предсказание сделать 2.5 ед инсулина! Сахар после этой компенсации стал = 200
            
            Выше нормы на 60! Мы вычеслаем корректировку correct_insulin - 0.6 И нам нужно 0.6 инсулина добавить
            к прошлой компенсации! Распределяем мы 0.6 в пропорции от массы углеводов! Винограду достанется 0.4
            а апельсину 0.2 Прибавка к дозе!
            
            Станет:
            
            Апельсин 100г Углеводы 10г - Инсулин под глюкозу 1.2
            Виноград 100г Углеводы 15г - Инсулин под глюкозу 1.9
            
            
            Возвращает список новых значений.
            
            '''
            
            sum_carbo =  df['Сумма углеводов'].values[0]

            df['Поправка инсулина'] = df['Углеводы'].apply(lambda x: x / sum_carbo * correct_insulin)

            # Таким образом если сахар выше нормы то за предыдущий обед мы повысим дозу инсулина!
            # Если ниже то понизим
            df['Новый инсулин под глюкозу'] = df['Инсулин под глюкозу'] + df['Поправка инсулина']
            
            # Вернем новый инсулин и сам коэффициент корректировки
            return df['Новый инсулин под глюкозу'].values, correct_insulin
    
    
    
    

        
```
