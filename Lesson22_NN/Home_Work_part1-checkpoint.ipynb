```python
# to make this notebook's output stable across runs
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
```

```python
from sklearn.cross_validation import train_test_split


X_train,X_valid,y_train,y_valid = train_test_split(X_moons_with_bias,y_moons.reshape(-1,1),test_size = 0.3,
                                                   random_state = 42)
```

```python
# Функция для получения рандомных данных
def random_batch(X_train, y_train, batch_size):
    rnd_indices = np.random.randint(0, len(X_train), batch_size)
    X_batch = X_train[rnd_indices]
    y_batch = y_train[rnd_indices]
    return X_batch, y_batch
```

```python
# Первая реализация это линейная регрессия только с функцией аткивации сигмоид
```

```python
reset_graph()

n_inputs = 2

X = tf.placeholder(tf.float32, shape=(None, n_inputs + 1), name="X")
y = tf.placeholder(tf.float32, shape=(None, 1), name="y")

# Случайная инициализация весов 3 - веса под каждый признак(с 1чкой)
theta = tf.Variable(tf.random_uniform([n_inputs + 1, 1], -1.0, 1.0, seed=42), name="theta")

# Скалярное произведение признаков на вес
logits = tf.matmul(X, theta, name="logits")

##### Если классификация то положим выводы на сигмоид! Чтобы определить класс! Функция активации
# y_proba = 1 / (1 + tf.exp(-logits))

y_proba = tf.sigmoid(logits)

##### Теперь нужно подсчитать ошибку! Есть функция ЛОГ ЛОССС

# epsilon = 1e-7  # to avoid an overflow when computing the log Улучшаем логарифм
# loss = -tf.reduce_mean(y * tf.log(y_proba + epsilon) + (1 - y) * tf.log(1 - y_proba + epsilon))

loss = tf.losses.log_loss(y, y_proba) 

##### Оптимизатор 

optimize = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(loss)
```

```python
init = tf.global_variables_initializer()

n_epochs = 1000
batch_size = 50
n_batches = int(np.ceil(m / batch_size))

with tf.Session() as sess:
    
    init.run()
    losses_epoch = []
    lossess_score = []

    for epoch in range(n_epochs):
        
        for batch_index in range(n_batches):
            
            X_batch, y_batch = random_batch(X_train, y_train, batch_size)
            sess.run(optimize, feed_dict={X: X_batch, y: y_batch})    
            
            
        loss_val = loss.eval({X: X_valid, y: y_valid})
        
            
        if epoch % 100 == 0:
            
            lossess_score.append(loss_val)
            losses_epoch.append(epoch)
            
            print("Epoch:", epoch, "\tLoss:", loss_val)
            
    # Координаты всех точек в области графика
    x_y = np.c_[xx.ravel(), yy.ravel()]
    # Добавляем отступ для нейронки
    X_new_with_bias = np.c_[np.ones([len(x_y), 1]), x_y] 
    
    
    # Считаем для валидационной выборки
    y_proba_val = y_proba.eval(feed_dict={X: X_valid, y: y_valid})
    
    # Считаем для всех точек для определения разделяющей
    y_proba_plot = y_proba.eval(feed_dict={X: X_new_with_bias})
    best_theta = theta.eval()
    
plt.plot(losses_epoch,lossess_score) 
plt.title('График функции')
plt.ylabel('Ось Y')
plt.xlabel('Ось X')
plt.grid(True)
plt.show()
```

```python
reset_graph()

n_neuros = 50
n_inputs = 2
n_class = 2

X = tf.placeholder(tf.float32, shape=(None, n_inputs + 1), name="X")
y = tf.placeholder(tf.float32, shape=(None, 1), name="y")

# реализуем архитекту  2 скрытых слоя по 50нейронов

with tf.name_scope('hidden'):
    
    
    # 2 полносвязных слоя с функцией активации elu! Сейчас одна из самых модных!
    hidden1 = tf.layers.dense(X, n_neuros, name = 'hidden1',activation=tf.nn.elu)
    hidden2 = tf.layers.dense(hidden1, n_neuros, name = 'hidden2',activation=tf.nn.elu)
    
    # собираем все значения в промежуточном слое
    logits = tf.layers.dense(hidden2, 1, name = 'logits')
    
with tf.name_scope('loss'):
    # вычесляем вероятности по результатм
    y_proba = tf.sigmoid(logits)
    
    # считаем ошибку!
    loss = tf.losses.sigmoid_cross_entropy(y,y_proba)
    


##### Оптимизатор 
with tf.name_scope('train'):
    
    optimize = tf.train.MomentumOptimizer(learning_rate=0.01, momentum=0.9).minimize(loss)

init = tf.global_variables_initializer()

n_epochs = 1000
batch_size = 50
n_batches = int(np.ceil(m / batch_size))

with tf.Session() as sess:
    # инициализирую переменные
    init.run()
    
    # собираю ошибку
    losses_epoch = []
    lossess_score = []

    for epoch in range(n_epochs):
        
        for batch_index in range(n_batches):
            
            X_batch, y_batch = random_batch(X_train, y_train, batch_size)
            sess.run(optimize, feed_dict={X: X_batch, y: y_batch})    
            
            
        loss_val = loss.eval({X: X_valid, y: y_valid})
        
            
        if epoch % 100 == 0:
            
            lossess_score.append(loss_val)
            losses_epoch.append(epoch)
            
            print("Epoch:", epoch, "\tLoss:", loss_val)
            
    # Координаты всех точек в области графика
    x_y = np.c_[xx.ravel(), yy.ravel()]
    # Добавляем отступ для нейронки
    X_new_with_bias_scaller = StandardScaler().fit_transform(X_new_with_bias)
    
    # Считаем для валидационной выборки
    y_proba_val = y_proba.eval(feed_dict={X: X_valid, y: y_valid})
    
    # Считаем для всех точек для определения разделяющей
    y_proba_plot = y_proba.eval(feed_dict={X: X_new_with_bias})
    
    
    
plt.plot(losses_epoch,lossess_score) 
plt.title('График функции')
plt.ylabel('Ось Y')
plt.xlabel('Ось X')
plt.grid(True)
plt.show()
```
