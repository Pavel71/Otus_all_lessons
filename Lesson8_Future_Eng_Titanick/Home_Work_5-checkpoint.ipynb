```python
from sklearn.preprocessing import  LabelEncoder, LabelBinarizer, OneHotEncoder
from sklearn.pipeline import make_union, make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, Imputer
from sklearn.preprocessing import CategoricalEncoder
from sklearn.base import BaseEstimator, TransformerMixin


class Simple_pipeline(BaseEstimator):
    
    def __init__(self, df):
        self.df = df
        
        # Embarked содержит 2 пропуска! Я решил заполнить их самым частым признаком
        # Возможно так делать не корректно но я хотел добавить именно строковый элемент,
        # чтобы можно было преобразовать сразу CategoricalEncoder

    def embarked_fill(df):
        max_count_embarked = df['Embarked'].value_counts().index[0]
        df['Embarked'] = df['Embarked'].fillna(max_count_embarked)
        return df

    # Объеденим родственников в 1 признак. просто сложим их

    def get_sex_embarked_pclass_col(df):
        # Pclass,SibSp,Parch уже числовые поэтому CategoricalEncoder для них сделает просто onehot
        return df[['Sex','Embarked','Pclass','SibSp','Parch']]

    def get_num_cols(df):
        return df[['Age', 'Fare']]
    
    def get_pipline(self):

        pipeline = make_union(*[
            make_pipeline(FunctionTransformer(embarked_fill, validate=False),
                          FunctionTransformer(get_sex_embarked_pclass_col, validate=False), 
                          CategoricalEncoder(encoding='onehot')), # Данных мало можно сделать полную матрицу 
                                                                        # для наглядности.

            make_pipeline(FunctionTransformer(get_num_cols, validate=False), Imputer(strategy='mean'),StandardScaler())
        ])
        
        return pipeline.fit_transform(self.df)
```

```python
from sklearn.preprocessing import  LabelEncoder, LabelBinarizer, OneHotEncoder
from sklearn.pipeline import make_union, make_pipeline
from sklearn.preprocessing import FunctionTransformer, StandardScaler, Imputer
from sklearn.preprocessing import CategoricalEncoder
from sklearn.base import BaseEstimator, TransformerMixin




class Full_pipeline(BaseEstimator):
    
    def __init__(self, df):
        self.df = df
        
        
        
    def name_future(self, df):
        
        df['Lname'] = df.Name.apply(lambda x: x.split(' ')[0].replace(',', ''))
        df['NamePrefix'] = df.Name.apply(lambda x: x.split(' ')[1])
        
        return df[['Lname','NamePrefix']]
        
    # Embarked содержит 2 пропуска! Я решил заполнить их самым частым признаком
    # Возможно так делать не корректно но я хотел добавить именно строковый элемент,
    # чтобы можно было преобразовать сразу CategoricalEncoder
        
    def embarked_fill(df):
        max_count_embarked = df['Embarked'].value_counts().index[0]
        df['Embarked'] = df['Embarked'].fillna(max_count_embarked)
        return df

    # Объеденим родственников в 1 признак. просто сложим их
    def family(self, df):
        
        Data['Family'] = np.sum(Data[['SibSp','Parch']],axis = 1)
        return Data['Family'].values.reshape(-1,1)
        

    def get_cat_col(self, df):
        # Pclass уже числовые поэтому CategoricalEncoder для них сделает просто onehot
        return df[['Sex','Embarked','Pclass']]

    def get_num_cols(df):
        return df[['Age', 'Fare']]
    
    # Признак категории возраста
    
    def get_age_cat(self, df):
        
        df['Age_cat'] = df['Age'].apply(lambda age: 1 if age < 30 
                                            else 3 if age > 60 else 2);
        return df['Age_cat'].values.reshape(-1,1)
    
    # Сделаем попарное взаимодействие признаков object
    
    def get_par_obj(self,df):
        
        df_interact = df[['Embarked','Sex','Lname','NamePrefix']]
        cat_features = df_interact.columns.values
        
        for i, col1 in enumerate(cat_features):
                
            for j, col2 in enumerate(cat_features[i + 1:]):
                    
                df_interact[col1 + '_' + col2] = df_interact[col1] + '_' + df_interact[col2]
                
        dfclean = df_interact.drop(['Embarked','Sex','Lname','NamePrefix'],axis = 1)        
                    
        return dfclean
    
    def get_pipline(self):

        pipeline = make_union(*[
            
            make_pipeline(FunctionTransformer(self.get_age_cat, validate=False),
                          CategoricalEncoder(encoding='onehot')),
            
            make_pipeline(FunctionTransformer(self.family, validate=False),
                          CategoricalEncoder(encoding='onehot')),
            
            make_pipeline(FunctionTransformer(self.name_future, validate=False), 
                          CategoricalEncoder(encoding='onehot')),
            
            make_pipeline(FunctionTransformer(embarked_fill, validate=False),
                          FunctionTransformer(self.get_cat_col, validate=False), 
                          CategoricalEncoder(encoding='onehot')),
            
#             make_pipeline(FunctionTransformer(self.get_par_obj, validate=False), 
#                           CategoricalEncoder(encoding='onehot')),
            
            make_pipeline(FunctionTransformer(get_num_cols, validate=False), Imputer(strategy='mean'),StandardScaler())
           
        ])
        
        return pipeline.fit_transform(self.df)
```
